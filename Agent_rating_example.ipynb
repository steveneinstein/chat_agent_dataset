```python
def convert_to_sec(var):
    H,M,S = list(map(int, var.split(':')))
    return (H * 60 * 60) + (M *60) + S
```

```python
k = '01:20:20'
k.split(':')
n = list(map(int ,(k.split(':'))))
n
```

```python
df['Average Response Time'] = df['Average Response Time'].apply(convert_to_sec)
```

```python
df['Average Resolution Time'] = df['Average Resolution Time'].apply(convert_to_sec)
```

```python
def check_ratings(var):
    if var > 4:
        return 'good'
    else:
        return 'need to improve'
```

```python
df['Ratings'] = df['Average Rating'].apply(check_ratings)
```
